```python
import pandas as pd
```

```python
belem_2020 = pd.read_csv("mesarios_belem_2020.csv", encoding="latin1")
belem_2024 = pd.read_csv("mesarios_belem_2024.csv", encoding="latin1")
poa_2020 = pd.read_csv("mesarios_poa_2020.csv", encoding="latin1")
poa_2024 = pd.read_csv("mesarios_poa_2024.csv", encoding="latin1")
```

```python
dataset = pd.concat([belem_2020, belem_2024, poa_2020, poa_2024], ignore_index=True)
```

```python
dataset
```

```python
dataset = dataset.drop(columns=["CD_MUNICIPIO", "CD_GENERO", "CD_ESTADO_CIVIL", "CD_FAIXA_ETARIA", "CD_GRAU_ESCOLARIDADE"])
```

```python
infos_dataset = dataset.describe(include="all")
```

```python
dataset.to_csv("dataset.csv", index=False, encoding='latin1')
```
